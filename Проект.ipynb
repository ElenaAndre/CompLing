```python
with open("По причине души.txt", "r") as open_file: 
    text = open_file.read()
print(text)
```

```python
!pip install nltk
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentence = sent_tokenize(text)
print(sentence)
```

```python
for i in sentence:
    print(i)
```

```python
print(type(sentence[0]))
```

```python
!pip install regex
import re
```

```python
target_sent = []
for i in sentence:
    result = re.search(r'[С,с]теп[ьиян].{0,2}|[С,с]тепей', i)
    if not result == None:
        print(result)
        target_sent.append(i)

for j in target_sent:
    print(j)
    
    

# print(result.group())
```
